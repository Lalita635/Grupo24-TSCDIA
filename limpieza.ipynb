# LIMPIEZA DE DATOS

## Celdas vacias

```python


#método isna(), comprobar si cada celda es nula y  sum(), suma los resultados
filas_vacias = df4.isna().sum()

# Ahora filas_vacias contiene la cantidad de valores nulos en cada columna
# Si deseas contar las filas vacías en su totalidad, puedes usar sum() nuevamente
total_filas_vacias = filas_vacias.sum()

print("Filas vacías por columna:")
print(filas_vacias)

print("Total de filas vacías en el DataFrame:")
print(total_filas_vacias)
```
